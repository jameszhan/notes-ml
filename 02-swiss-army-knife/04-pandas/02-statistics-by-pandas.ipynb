```python
print("data.shape = {0}".format(data.shape))
print("data.columns = {0}".format(data.columns))
print("data['Gender'] = {0}".format(data['Gender']))
print("data[data['Gender'] == 'Female']['VIQ'].mean() = {0}".format(data[data['Gender'] == 'Female']['VIQ'].mean()))
print("data[data['Gender'] == 'Male']['VIQ'].mean() = {0}".format(data[data['Gender'] == 'Male']['VIQ'].mean()))
```

```python
groupby_gender = data.groupby('Gender')
for gender, value in groupby_gender['VIQ']:
    print((gender, value.mean()))

groupby_gender.mean()
```
