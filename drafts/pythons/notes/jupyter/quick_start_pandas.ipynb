```python
print("df.head() = \n{0}\n".format(df.head()))
print("df.tail() = \n{0}\n".format(df.tail(3)))
print("df.index = \n{0}\n".format(df.index))
print("df.columns = \n{0}\n".format(df.columns))
print("df.values = \n{0}\n".format(df.values))
print("df.describe() = \n{0}\n".format(df.describe()))
```

```python
print("df.T = \n{0}\n".format(df.T))
print("df.sort_index(axis=0, ascending=False) = \n{0}\n".format(df.sort_index(axis=0, ascending=False)))
print("df.sort_index(axis=1, ascending=False) = \n{0}\n".format(df.sort_index(axis=1, ascending=False)))
print("df.sort_values(by='B') = \n{0}\n".format(df.sort_values(by="B")))
```

```python
print("df['B'] = \n{0}\n".format(df['B']))
print("df.B = \n{0}\n".format(df.B))
print("df[1:3] = \n{0}\n".format(df[1:3]))
print("df['2018-01-02':'2018-01-06'] = \n{0}\n".format(df['2018-01-02':'2018-01-06']))
```

```python
print("df.loc['2018-01-01'] = \n{0}\n".format(df.loc['2018-01-01']))
print("df.loc[dates[0]] = \n{0}\n".format(df.loc[dates[0]]))
print("df.loc[:,['A','B']] = \n{0}\n".format(df.loc[:,['A','B']]))
print("df.loc['20180102':'20180106',['A','B']] = \n{0}\n".format(df.loc['20180102':'20180106',['A','B']]))
print("df.loc['2018-01-01',['A','B']] = \n{0}\n".format(df.loc['2018-01-01',['A','B']]))
print("df.loc['2018-01-01','B'] = \n{0}\n".format(df.loc['2018-01-01','B']))
```

```python
print("dates[0] = {0}".format(dates[0]))
print("df.at['2018-01-01','B'] = \n{0}\n".format(df.at[dates[0],'B']))
%timeit -n10000 df.loc[dates[0],'B']
%timeit -n10000 df.at[dates[0],'B']
```

```python
print("df.iloc[3] = {0}".format(df.iloc[3]))
print("df.iloc[3:5,0:2] = \n{0}\n".format(df.iloc[3:5,0:2]))
print("df.iloc[[1,2,4],[0,2]] = \n{0}\n".format(df.iloc[[1,2,4],[0,2]]))
print("df.iloc[1:3,:] = \n{0}\n".format(df.iloc[1:3,:]))
print("df.iloc[:, 1:3] = \n{0}\n".format(df.iloc[:, 1:3]))

%timeit -n100 df.iloc[1,1]
%timeit -n100 df.iat[1,1]
```

```python
print("df[df.A > 0] = \n{0}\n".format(df[df.A > 0]))
print("df[df > 0] = \n{0}\n".format(df[df > 0]))
```

```python
df3 = df.copy()
df3.at['2018-01-01', 'A'] = 10
print("df3 = \n{0}\n".format(df3))
df3.loc[:,'D'] = [5] * len(df)
print("df3 = \n{0}\n".format(df3))
df3.loc['2018-01-02':'2018-01-06','B'] = None
print("df3 = \n{0}\n".format(df3))
print("pd.isnull(df3) = \n{0}\n".format(pd.isnull(df3)))
print("df3.fillna(value=99) = \n{0}\n".format(df3.fillna(value=99)))
print("df3.dropna(how='any') = \n{0}\n".format(df3.dropna(how='any')))
```

```python
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three',
                          'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})
print("df = \n{0}\n".format(df))
print("df.groupby('A').sum() = \n{0}\n".format(df.groupby('A').sum()))
print("df.groupby(['A', 'B']).sum() = \n{0}\n".format(df.groupby(['A', 'B']).sum()))
```
